## Reshaping, stacking , squeezing and unsqueezing tensors
* Reshaping - reshapes an input to a defined shape
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - combine the

```python
# Create a tensor
import torch
x = torch.arange(1., 8.)
x, x.shape
```

```python
# Now let's add an extra dimension with torch.reshape()
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

```python
# Change the view(keeps same data as original but changes the view)
z = x.view(1, 7)
z, z.shape
```

```python
# Changing z changes x too since they share the same memory location
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)  # try changing dim to dim=1 and see what happens
x_stacked
```

```python
# squeezing -> removing all single dimensions from a tensor using  torch.squeeze()
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New Shape: {x_squeezed.shape}")
```

```python
# And to do the reverse of torch.squeeze() you can use torch.unsqueeze() to add a dimenson value of 1 at a specific index
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")


# Add an extra dimension with unsqueezed
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"\nNew shape: {x_unsqueezed.shape}")
```

```python
"""You can also rearrange the order of axes values with torch.permute(input, dims),
where the input gets turned into a view with new dims"""
# Create tensor with specific shape

x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1)  # shifts axis 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"\nNew shape: {x_permuted.shape}")
```


## Pytorch tensors and Numpy
* torch.from_numpy(ndarray) - Numpy array -> Pytorch tensor.
* torch.Tensor.numpy() - Pytorch tensor -> NumPy array.




```python
# Numpy array to tensor
import torch
import numpy as np
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# convert Numpy array(float64) -> Pytorch tensor(float64) -> pytorch tensor (float32)
tensor_pytorch_float32 = tensor
tensor_pytorch_float32 = torch.from_numpy(array).type(torch.float32)
tensor_pytorch_float32, tensor_pytorch_float32.dtype
```

```python
# change the array , keep the tensor
array = array + 1
array, tensor
```

```python
# PyTorch to Numpy array
tensor = torch.ones(7)  # create a tensor of ones with dtype=float32
numpy_tensor = tensor.numpy()  # will be dtype=float32 unless changed
tensor, numpy_tensor
```

```python
# Change the tensor, keep the array the same
tensor = tensor + 1
tensor, numpy_tensor
```

## Reproducibility (Trying to take the random out of random)

```python
import torch
# create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)


print(f"Tensor A:\n{random_tensor_A}\n")
print(f"Tensor B:\n{random_tensor_B}\n")
print(f"Does Tensor A equal Tensor B?(anywhere)\n")
random_tensor_A == random_tensor_B
```

```python
# torch.manual_seed(seed) for creating random tensors with the same flavour
import torch
import random

# set the random seed
RANDOM_SEED=42 # Try changing this to random values and see what happens to the numbers below
torch.manual_seed(seed=RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

# Have to reset the seed every time a new rand() is called
# Without this, tensor_D, would be different to tensor_C
torch.random.manual_seed(seed=RANDOM_SEED)  # try commenting ths line and see what happens
random_tensor_D = torch.rand(3, 4)

print(f"Tensor C:\n{random_tensor_C}\n")
print(f"Tensor D:\n{random_tensor_D}\n")
print(f"Does Tensor C equal Tensor D (anywhere)")
random_tensor_C == random_tensor_D
```

```python
# Getting pytorch to run on the GPU
import torch
torch.cuda.is_available()
```

```python
# set the device type
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# Count number of devices
torch.cuda.device_count()
```

```python
# Putting tensors (and models) on the GPU
# Create tensor (default on CPU)
tensor = torch.tensor([1, 2, 3])

# Tensor not on GPU
print(tensor, tensor.device)

# Move tensor to GPU (if available)
tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

```python
# Moving tensors back to the CPU

# If tensor is on GPU, can't transform it to numpy (this will error)
tensor_on_gpu.numpy()
```

```python
# Instead to get a tensor back to cpu and useable with Numpyy we can use Tensor.cpu()
# This copies the tensor to CPU memory so it's usable with CPU

# Instead, copy the tensor back to cpu
tensor_back_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_back_on_cpu
```
